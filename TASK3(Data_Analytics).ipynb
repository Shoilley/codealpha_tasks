```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load sample dataset
df = sns.load_dataset('tips')  # Dataset about restaurant tips

# Step 3: Display the first few rows
df.head()

# Step 4: Visual 1 - Total bill vs Tip
plt.figure(figsize=(8, 5))
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plt.title('Total Bill vs Tip')
plt.show()

# Step 5: Visual 2 - Distribution of Total Bill
plt.figure(figsize=(8, 5))
sns.histplot(df['total_bill'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Total Bill')
plt.show()

# Step 6: Visual 3 - Average Tip by Day
plt.figure(figsize=(8, 5))
sns.barplot(x='day', y='tip', data=df, estimator='mean', ci=None)
plt.title('Average Tip by Day')
plt.show()
```
